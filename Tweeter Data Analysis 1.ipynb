```python
import pandas as pd
import matplotlib.pyplot as plt
import io
import base64

# Read data from CSV file
df = pd.read_csv('merged_created_at.csv') 

# Convert created_at column to datetime
df['created_at'] = pd.to_datetime(df['created_at'])

# Filter for weekday/weekend
weekdays = df[(df['created_at'].dt.weekday < 5)]  
weekends = df[(df['created_at'].dt.weekday >= 5)]

# Count tweets by hour for weekdays and weekends  
tweets_weekdays = weekdays.groupby(weekdays['created_at'].dt.hour).size()
tweets_weekends = weekends.groupby(weekends['created_at'].dt.hour).size()

# Plot average tweets by hour for weekdays and weekends
fig, ax = plt.subplots(figsize=(10,8))
tweets_weekdays.groupby(tweets_weekdays.index).mean().plot(label='Weekdays')
tweets_weekends.groupby(tweets_weekends.index).mean().plot(label='Weekends')
ax.set_title('Average tweets by hour')
ax.set_xlabel('Hour')
ax.set_ylabel('Average number of tweets')
ax.legend()

# Save the plot as an image file
buffer = io.BytesIO()
plt.savefig(buffer, format='png')
buffer.seek(0)
image_png = buffer.getvalue()
buffer.close()

# Encode the image as base64
image_base64 = base64.b64encode(image_png).decode()

# Create download link
download_link = f'<a href="data:image/png;base64,{image_base64}" download="tweets_by_hour.png">Download Plot</a>'

# Display download link
display(HTML(download_link))
```

```python
import os
import json
import pandas as pd

# Function to extract country from a line in a JSON file
def extract_countries_from_file(file_path):
    countries = []
    with open(file_path, 'r', encoding='utf-8') as json_file:
        for line in json_file:
            try:
                data = json.loads(line)
                if isinstance(data, dict):
                    country = data.get('place', {}).get('country')
                    if country:
                        countries.append(country)
            except (json.JSONDecodeError, AttributeError):
                continue
    return countries

# Initialize an empty list to store extracted data
data_list = []

# Define the base directory
base_directory = '/kaggle/input/laurenbux/TwitterJune2022'

# Walk through the base directory and its subdirectories
for root, dirs, files in os.walk(base_directory):
    for file in files:
        if file.endswith('.json'):
            file_path = os.path.join(root, file)
            countries = extract_countries_from_file(file_path)
            for country in countries:
                data_list.append({'file_path': file_path, 'country': country})

# Create a DataFrame from the extracted data
df = pd.DataFrame(data_list)

# Display the first few rows to check the 'country' column
print(df.head())
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import os
import json

# Function to read JSON files and combine data into a DataFrame
def read_json_files(base_folder):
    dataframes = []
    for root, dirs, files in os.walk(base_folder):
        for file in files:
            if file.endswith(".json"):
                file_path = os.path.join(root, file)
                with open(file_path, 'r', encoding='utf-8') as file:
                    for line in file:
                        try:
                            json_data = json.loads(line)
                            df = pd.DataFrame([json_data])
                            dataframes.append(df)
                        except json.JSONDecodeError as e:
                            print(f"Error reading JSON in file {file_path}: {e}")
                            continue
    if dataframes:
        return pd.concat(dataframes, ignore_index=True)
    else:
        return pd.DataFrame()

# Replace 'base_folder' with the path to your base folder containing JSON files
base_folder_path = '/kaggle/input/laurenbux/TwitterJune2022'

# Read JSON files into a DataFrame
combined_df = read_json_files(base_folder_path)

if not combined_df.empty:
    # Convert 'created_at' field to datetime
    combined_df['created_at'] = pd.to_datetime(combined_df['created_at'])

    # Save the dataframe to HDF5 file
    combined_df.to_hdf('combined_df.h5', 'df')

    # Load the dataframe from HDF5 file
    combined_df = pd.read_hdf('combined_df.h5', 'df')

    # Find the day with unusually high activity
    activity_by_day = combined_df['created_at'].dt.date.value_counts()
    most_active_day = activity_by_day.idxmax()
    print(f"The day with the highest activity is: {most_active_day}")

    # Filter data for the most active day
    most_active_tweets = combined_df[combined_df['created_at'].dt.date == most_active_day]

    # Generate a word cloud directly from tweet texts
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(most_active_tweets['text']))

    # Display the word cloud
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud of Tweets on the Most Active Day')
    plt.show()
else:
    print("No JSON files found in the specified folder path.")
```

```python
import os
import json

# Define your base folder containing subdirectories with JSON files
base_folder = '/kaggle/input/laurenbux/TwitterJune2022'

# Function to extract tweets from a file for a specific date
def extract_tweets_from_file(file_path, target_date):
    tweets = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            tweet = json.loads(line)
            created_date = tweet.get('created_at', '')[:10]  # Extract date part
            if created_date == target_date:
                tweets.append(tweet)
    return tweets

target_date = '2022-06-01'
tweets_for_date = []

# Walk through all files in the base folder and its subdirectories
for root, dirs, files in os.walk(base_folder):
    for file_name in files:
        if file_name.endswith('.json'):
            file_path = os.path.join(root, file_name)
            tweets_for_date.extend(extract_tweets_from_file(file_path, target_date))

# 'tweets_for_date' now contains tweets from the specified date across all files
print(f"Total tweets on {target_date}: {len(tweets_for_date)}")
```
